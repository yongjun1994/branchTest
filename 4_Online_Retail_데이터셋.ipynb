```python
# 비회원/탈퇴/휴면회원 제거
retail = retail[pd.notnull(retail['Customer ID'])] # retail이라는 데이터프레임에서 'Customer ID'값이 결측값이 아닌 행만 필터링
retail
```

```python
len(retail)
```

```python
# 구입 수량이 0 이하인 데이터를 확인
retail[retail['Quantity'] <= 0]
```

```python
# 구입 수량이 1 이상인 데이터만 저장
retail = retail[retail['Quantity'] >= 1]
len(retail)
```

```python
# 구입 가격이 0이하인 데이터를 확인
retail[retail['Price'] <= 0]
```

```python
# 구입 가격이 0보다 큰 데이터만 저장
retail = retail[retail['Price'] > 0]
len(retail)
```

```python
# 고객의 총 지출비용(CheckoutPrice) 파생변수 만들기
# 총 지출비용(CheckoutPrice) = 가격(Price) + 수량(Quantity)
retail['CheckoutPrice'] = retail['Price'] * retail['Quantity']
retail
```

```python
retail.info()
```

```python
# 날짜 형식으로 바꾸어 재저장
retail['InvoiceDate'] = pd.to_datetime(retail['InvoiceDate'])
retail.info()
```

```python
# 전체 매출
total_revenue = retail['CheckoutPrice'].sum()
total_revenue
```

```python
# 월별 매출 구하기
retail['InvoiceDate'].sort_values(ascending=False)
```

```python
def extract_month(date): #  date에는 날짜가 들어옴
    month = str(date.month) # 1
    if date.month < 10:
        month = '0' + month
    return str(date.year) + month # 201101
```

```python
# # 데이터프레임에서 'InvoiceDate' 열을 인덱스로 설정
# # 'InvoiceDate'는 거래 날짜를 나타내는 열로, 이를 인덱스로 설정하면 날짜별로 데이터 정렬 및 그룹화가 가능해짐
# retail.set_index('InvoiceDate')

# # 데이터를 월 단위로 그룹화
# # 'extract_month'는 각 거래 날짜에서 월(month)을 추출하는 함수 또는 열
# retail.set_index('InvoiceDate').groupby(extract_month)

# # 'CheckoutPrice' 열의 합계를 계산
# # 각 월(extract_month)별로 'CheckoutPrice'(거래 금액)의 총합을 계산
# retail.set_index('InvoiceDate').groupby(extract_month)['CheckoutPrice'].sum()

# # 최종 결과를 변수 'rev_by_month'에 저장
# # 'rev_by_month'는 월별 총 매출 데이터를 담고 있음
# rev_by_month = retail.set_index('InvoiceDate').groupby(extract_month)['CheckoutPrice'].sum()
```

```python
# 월별 checkoutprice 구함
rev_by_month = retail.set_index('InvoiceDate').groupby(extract_month)['CheckoutPrice'].sum()
rev_by_month
```

```python
# Figure와 Axes 생성
# figsize=(20, 10): 그래프의 크기를 가로 20인치, 세로 10인치로 설정
fig, ax = plt.subplots(figsize=(20, 10))

# 막대 그래프 그리기
# x축: rev_by_month의 인덱스 (나라 이름)
# y축: rev_by_month의 값 (각 나라의 매출)
ax.bar(rev_by_month.index, rev_by_month.values)

# x축 레이블 설정
# 레이블 텍스트: 'Country', 글씨 크기: 12
ax.set_xlabel('Country', fontsize=12)

# y축 레이블 설정
# 레이블 텍스트: 'Revenue', 글씨 크기: 12
ax.set_ylabel('Revenue', fontsize=12)

# 그래프 제목 설정
# 제목 텍스트: 'Revenue by Country', 글씨 크기: 15
ax.set_title('Revenue by Country', fontsize=15)

# x축 눈금 위치 설정
# rev_by_month.index의 크기만큼 눈금을 추가
ax.set_xticks(rev_by_month.index)

# x축 눈금 레이블 설정
# x축 레이블을 rev_by_month의 인덱스 값으로 설정하고, 각 레이블을 45도 회전
ax.set_xticklabels(rev_by_month.index, rotation=45)

# 레이아웃 조정
# tight_layout()으로 그래프 요소 간 간격을 자동으로 조정하여 보기 좋게 배치
plt.tight_layout()

# 그래프를 화면에 출력
plt.show()
```

```python
def plot_bar(df, xlabel, ylabel, title, figsize=(20, 10), fontsize=12, titlesize=15, rotation=45):
    fig, ax = plt.subplots(figsize=figsize)
    ax.bar(df.index, df.values)
    ax.set_xlabel(xlabel, fontsize=fontsize)
    ax.set_ylabel(ylabel, fontsize=fontsize)
    ax.set_title(title, fontsize=titlesize)
    ax.set_xticks(range(len(df.index)))
    ax.set_xticklabels(df.index, rotation=rotation, fontsize=fontsize)
    plt.tight_layout()
    plt.show()
```

```python
plot_bar(rev_by_month, "Month", "Revenue", "Revenue By Month")
```

```python
# 요일별 매출 구하기
def extract_dow(date):
    return date.dayofweek
```

```python
# # 'InvoiceDate' 열을 인덱스로 설정
# # 'InvoiceDate'는 거래 날짜를 나타내는 열로, 이를 인덱스로 설정하여 날짜 기반 그룹화를 가능하게 함
# retail.set_index('InvoiceDate')

# # 데이터를 요일(Day of Week) 기준으로 그룹화
# # lambda date: date.dayofweek는 각 날짜(date)에서 요일을 추출하는 람다 함수
# # 요일 값은 0(월요일)부터 6(일요일)까지 정수로 반환됨
# retail.set_index('InvoiceDate').groupby(lambda date: date.dayofweek)

# # 'CheckoutPrice' 열의 합계를 계산
# # 각 요일별로 'CheckoutPrice'(거래 금액)의 총합을 계산
# retail.set_index('InvoiceDate').groupby(lambda date: date.dayofweek)['CheckoutPrice'].sum()

# # 결과를 변수 'rev_by_dow'에 저장
# # 'rev_by_dow'는 요일별 총 매출 데이터를 담고 있음
# rev_by_dow = retail.set_index('InvoiceDate').groupby(lambda date: date.dayofweek)['CheckoutPrice'].sum()
```

```python
# rev_by_dow = retail.set_index('InvoiceDate').groupby(extract_dow)['CheckoutPrice'].sum()
rev_by_dow = retail.set_index('InvoiceDate').groupby(lambda date: date.dayofweek)['CheckoutPrice'].sum()
rev_by_dow
```

```python
DAY_OF_WEEK = np.array(['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun'])
rev_by_dow.index = DAY_OF_WEEK[rev_by_dow.index]
rev_by_dow.index
```

```python
plot_bar(rev_by_dow, "DOW", "Revenue", "Revenue By Dow")
```
